## Part 6: Linear LSE
### (35 marks)

### Curve Fitting

curvefitting.png

### Observations: values of a function f(a) in range of - to , contaminated with random noise

### Unknown States: coefficients of polynomial functions for curve fitting

```python
# step 1
# import necessary libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import requests
```

```python
# step 2
# Load data.xlsx file, which contains the observations (with noise), variance of observations and true observations
# hint: use pandas library to parse excel file and use numpy library to convert it to array
# (1 mark)

# Download Excel File
url = 'https://github.com/nathanielmills-101/ENGG680_2025_Fall/raw/main/Lab_Assignment%201/data.xlsx'
r = requests.get(url)
with open('data.xlsx', 'wb') as f:
    f.write(r.content)

# Read Excel
df = pd.read_excel('data.xlsx')

# Clean spaces in the columns names
df.columns = df.columns.str.strip()

# Extracting Numpy Arrays using the correct name
y_obs = df['l'].to_numpy()       # Noise observations
variance = df['l_var'].to_numpy()  # Variance of Observations
y_true = df['l_true'].to_numpy()   # True observations

# Generate X as index row
x = np.arange(len(y_obs))

# Print first rows to verify
print("x:", x[:5])
print("y_obs:", y_obs[:5])
print("variance:", variance[:5])
print("y_true:", y_true[:5])
```

```python
# step 3
# define the number of observation n
# (1 mark)
n = len(y_obs)

# define the observation weight matrix P for LSE as identity matrix
# identity weight matrix simply means the observations will be unweighted in the following LSE
# (1 mark)
P = np.identity(n)

# define the range of function input(i.e., constants a), from start to end; and the numerical array of a
start = -1* np.pi
end = np.pi
a = np.linspace(start, end, n)
```

```python
# step 4
# use the matplotlib to plot the true and noisy observations together in one single figure
# use scatter for noisy observation plotting
# for the true observation, label it as "f(a)"" and set color as 'k'
# for the noisy observation scatter plot, label it as "f(a) + e", set color as 'tab:blue', set marker as 'o' and set alpha as 0.3
# include proper axis labels, legend and title
# scale the range of x-axis to be -1 to 1
# keep this format for the rest of the plots in this lab assignment
# (2 marks)

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(14, 5))

# Graphic of True Observations
ax.plot(a, y_true, label='f(a)', color='k')

# Graphic of Noisy Observations
ax.scatter(a, y_obs, label='f(a) + e', color='tab:blue', marker='o', alpha=0.3)

# Axis labes, legend and title
ax.set_xlabel('a (radians)')
ax.set_ylabel('f(a) + e')
ax.set_title('True and Noisy Observations')

#Labels and limits of x-axis
ax.legend()
ax.set_xlim(-1,1)

#Format Adjustment
ax.grid(True, which='both',ls='-',color='0.65')
fig.tight_layout()
```

```python
# step 5
# define a curve fitting function named fit_polynomial_curve
# the function should implement the linear least square estimation
# the function should include all important intermediate steps in LSE
# the input and output of the funciton are described in the comment below in red.
# please keep code clean and well-commented
# (7 marks)

def fit_polynomial_curve(a: np.ndarray, l: np.ndarray, P: np.ndarray, deg: int) -> tuple:
    '''
    Args:
        a:              function input (constants)
        l:              noisy observations
        P:              observation weight matrix
        deg:            degree of polynomial to fit to data
    Returns:
        x:              estimated polynomial coefficients
        cov_x:          uncertainty in estimated coefficients
        apv:            a posterier variance factor (goodness of fit)
        v:              observation residuals
    '''

    # Matrix A
    A = np.vander(a, deg + 1, increasing=True)

    # Estimaed coeficients X
    N = A.T @ P @ A
    n_inv = np.linalg.inv(N)
    x = n_inv @ (A.T @ P @ l)

    # Calculate the residuals
    v = A @ x - l

    # APV
    n = len(l)
    m = deg + 1
    apv = (v.T @ P @ v) / (n - m)

    # Covariance of Coeficients
    cov_x = apv * n_inv

    #Result return
    return (x, cov_x, apv, v)
```

```python
# step 6
# call the fit_polynomial_curve function and fetch the results
# use polynomial degrees of: 2, 4, 6
# (1 marks)

#Polynomial adjusment
x_deg2, cov_deg2, apv_deg2, v_deg2 = fit_polynomial_curve(a, y_obs, P, deg=2)
x_deg4, cov_deg4, apv_deg4, v_deg4 = fit_polynomial_curve(a, y_obs, P, deg=4)
x_deg6, cov_deg6, apv_deg6, v_deg6 = fit_polynomial_curve(a, y_obs, P, deg=6)

#Print results
print("=== Polynomial Adjustment ===")
print("Grade 2 -> Coeficients:", x_deg2)
print("Grade 2 -> Posterior Variance:", apv_deg2, "\n")

print("Grade 4 -> Coeficients:", x_deg4)
print("Grade 4 -> Posterior Variance:", apv_deg4, "\n")

print("Grade 6 -> Coeficients:", x_deg6)
print("Grade 6 -> Posterior Variance:", apv_deg6, "\n")
```

```python
# step 12
# now, let's define the P matrix with the proper weights of observation
# use the variance of observation l_var, define P matrix
# (1 mark)

# Define the P matrix using observation variances
P = np.diag(1 / df['l_var'].values)

print("P matrix (first 5 rows and columns):")
print(P[:5, :5]) # Showing the first 5 rows of the Matrix
```

```python
# Step 13
# Repeat Steps 6-10 using the proper P matrix
# (7 marks)

import numpy as np

# Store results
coef_list = []          # coefficients for each polynomial
y_pred_list = []        # predicted values
posterior_var_list = [] # posterior variance
norm_list = []          # L2 norm of residuals

# Observations
y = df['l'].values
x_obs = df['l_true'].values  # independent variable
poly_degrees = [1, 2, 3, 4, 5, 6]
max_degree = max(poly_degrees)

# Construct full Vandermonde matrix (design matrix) up to max degree
X = np.vander(x_obs, N=max_degree+1, increasing=True)

# Loop through polynomial degrees
for deg in poly_degrees:
    # Construct design matrix for this degree
    X_deg = X[:, :deg+1]

    # Weighted Least Squares: beta = (X^T P X)^-1 X^T P y
    XtPX = X_deg.T @ P @ X_deg
    XtPy = X_deg.T @ P @ y
    beta = np.linalg.inv(XtPX) @ XtPy

    # Store coefficients
    coef_list.append(beta)

    # Predicted values
    y_pred = X_deg @ beta
    y_pred_list.append(y_pred)

    # Posterior variance factor
    n = len(y)
    m = deg + 1
    residuals = y - y_pred
    sigma_hat2 = (residuals.T @ P @ residuals) / (n - m)
    posterior_var_list.append(sigma_hat2)

    # L2 norm of residuals
    norm_val = np.linalg.norm(residuals)
    norm_list.append(norm_val)

# Print results
for i, deg in enumerate(poly_degrees):
    print(f"\nPolynomial degree: {deg}")
    print(f"Coefficients: {coef_list[i]}")
    print(f"Posterior variance factor: {posterior_var_list[i]}")
    print(f"L2 norm of residuals: {norm_list[i]}")
```
